#6 Escala los datos, si consideras necesario


```python
from sklearn.preprocessing import StandardScaler

X = df.loc[:, df.columns != 'Y'].values 
Y = df['Y'].values 

xscaled = StandardScaler().fit_transform(X) 

scaledDf = pd.DataFrame(xscaled) 
scaledDf
```

# 7 Reduce las dimensiones con PCA, si consideras necesario.

a. Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploraciÃ³n de los datos la varianza > 70%
b. Indica la importancia de las variables en cada componente

```python
from sklearn.decomposition import PCA

pcaDf = PCA(n_components = 10) 
componentes = pcaDf.fit_transform(xscaled)
print(componentes.shape)
 
df_con_pca = pd.DataFrame( data=componentes, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10'])

target_names = {0.0:'Ladron',1.0:'No ladron'}
 
df_con_pca['Y'] = Y 
df_con_pca['Y'] = df_con_pca['Y'].map(target_names)
 
df_con_pca.head()

print(pcaDf.explained_variance_ratio_)
pcaDf.explained_variance_ratio_.sum()
df_con_pca
```

# 9 Realiza la visualizaciÃ³n de los datos usando por lo menos 3 grÃ¡ficos que consideres adecuados: plot, scatter, jointplot, boxplot, areaplot, pie chart, pairplot, bar chart, etc.


```python
sns.set()
 
plt.bar(
    range(1,len(pcaDf.explained_variance_)+1),
    pcaDf.explained_variance_
    )
 
plt.plot(
    range(1,len(pcaDf.explained_variance_)+1),
    np.cumsum(pcaDf.explained_variance_),
    c='red',
    label='Sumatoria de varianza explicada')
 
plt.legend(loc='upper left')
plt.xlabel('Number of components')
plt.ylabel('Explained variance (eignenvalues)')
plt.title('Scree plot')
 
plt.show()
```

```python
sns.jointplot(x = "PC1", y = "PC2", kind = "scatter", data = df_con_pca)
plt.show()
```

```python
pca = PCA(n_components=2)

X = df.loc[:, df.columns != 'Y'].values
Y = df['Y'].values 
X_escalada = StandardScaler().fit_transform(X)

columnas_pca = pca.fit_transform(X_escalada)
 
pca_df = pd.DataFrame(
    data=columnas_pca, 
    columns=['PC1', 'PC2'])
 
target_names = {
    0.0:'NOT_CREDIBLE_CLIENT',
    1.0:'CREDIBLE_CLIENT'
}
 
pca_df['target'] = Y
pca_df['target'] = pca_df['target'].map(target_names)
 
sns.set()
 
sns.lmplot(
    x='PC1', 
    y='PC2', 
    data=pca_df, 
    hue='target', 
    fit_reg=False, 
    legend=True
    )
 
plt.title('2D PCA Graph')
plt.show()
```
